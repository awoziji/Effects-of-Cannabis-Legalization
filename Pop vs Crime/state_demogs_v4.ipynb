```python
def total_pop(df, area):
    
    try:
        # AGEGRP 0 is total
        tot = df.loc[df['AGEGRP'] == 0, ]

        tot_group_year = tot.groupby('YEAR')
        tot_group_year = tot_group_year.sum()
        tot_group_year.reset_index(inplace=True)
        return tot_group_year['TOT_POP']

        #tot_group_year.plot('YEAR', 'TOT_POP', figsize=(18,9))
        #plt.title(f'Total population of {area}')
    except:
        return f'TypeError for {area}!'
```

```python
def tot_crime(inpt, crime, high, low):
    if type(inpt) == dict:
        dictionary = crime_add(inpt, high, low)
        output = []
        states = []
        for i in range(3, 10):
            output.append([])
            for key in dictionary:
                index = i - 3
                loop_series = dictionary[f'{key}'].loc[dictionary[f'{key}']['YEAR'] == i, : ]
                loop_series = loop_series.groupby('CTYNAME').sum()[f'{crime}'] / 19
                loop_total = loop_series.sum()
                output[index].append(loop_total)
                if i == 3:
                    states.append(key)

        for i in range(3, 10):
            index = i - 3
            loop_sum = 0

            for j in output[index]:
                loop_sum = loop_sum + j

            output[index] = loop_sum / len(output[index])
            
        return output
            
'''        fig, ax2 = plt.subplots()

        print(f'{crime} in {states}')
        
        avg_pop_from_dict(inpt, high, low)

        
        ax2.plot(['2010', '2011', '2012', '2013', '2014', '2015', '2016'], output)'''
```

<h1> Statewide </h1>

```python
def final_pop_state(dictionary):
    for key in dictionary:
        y_vals = dictionary[key]
        x_vals = ['2010','2011','2012','2013','2014','2015','2016']

        plt.figure(figsize=(18,9))

        plt.plot(x_vals, y_vals, color = 'forestgreen', linewidth = 4, zorder=1)
        plt.title(f'Statewide Population in {key} (2010-2016)', loc = 'center',fontsize = 17,fontweight = 3, color = 'black')
        plt.xlabel('Year' , fontsize = 15, fontweight = 3, color = 'black' )
        plt.ylabel('Population', fontsize = 15, fontweight= 3, color = 'black' )
        sns.set(rc={'axes.facecolor':'whitesmoke', 'figure.facecolor':'lightgrey'})
        plt.grid(color='gainsboro')
        plt.xlim(2010,2016)

        plt.savefig(f'output_graphs/state_wide/tot_pop/{key}_state_pop')
```
