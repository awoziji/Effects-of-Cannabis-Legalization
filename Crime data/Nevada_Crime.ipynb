```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2010 data
nv_10_path = "County_data/Nevada_county_2010.csv"
nv_10 = pd.read_csv(nv_10_path)
nv_10 = nv_10[['County','Violent crime']]
nv_10.head()
```

```python
# 2011 data
nv_11_path = "County_data/Nevada_county_2011.csv"
nv_11 = pd.read_csv(nv_11_path)
nv_11 = nv_11[['County','Violent crime']]
nv_11.head()
```

```python
# 2012 data
file_path = "County_data/Nevada_county_2012.csv"
nv_12 = pd.read_csv(file_path)
nv_12 = nv_12[['County','Violent crime']]
nv_12.head()
```

```python
# 2013 data
file_path = "County_data/Nevada_county_2013.csv"
nv_13 = pd.read_csv(file_path)
nv_13 = nv_13[['County','Violent crime']]
nv_13.head()
```

```python
# 2014 data
file_path = "County_data/Nevada_county_2014.csv"
nv_14 = pd.read_csv(file_path)
nv_14 = nv_14[['County','Violent crime']]
nv_14.head()
```

```python
# 2015 data
file_path = "County_data/Nevada_county_2015.csv"
nv_15 = pd.read_csv(file_path)
nv_15 = nv_15[['County','Violent crime']]
nv_15.head()
```

```python
# 2016 data
file_path = "County_data/Nevada_county_2016.csv"
nv_16 = pd.read_csv(file_path)
nv_16 = nv_16[['County','Violent crime']]
nv_16.head()
```

```python
# Merging 2010-2011 data
result_1 = pd.merge(nv_10, nv_11, how='left', on=['County'])
result_1 = result_1.rename(columns={'Violent crime_x':'2010','Violent crime_y':'2011'})
result_1.head()
```

```python
# Merging with 2012 data
result_2 = pd.merge(result_1, nv_12, how='left', on=['County'])
result_2.head()
```

```python
# Merging with 2013 data
result_3 = pd.merge(result_2, nv_13, how='left', on=['County'])
result_3 = result_3.rename(columns={'Violent crime_x':'2012','Violent crime_y':'2013'})
result_3.head()
```

```python
# Merging with 2014 data 
result_4 = pd.merge(result_3, nv_14, how='left', on=['County'])
result_4.head()
```

```python
# Merging with 2015 data
result_5 = pd.merge(result_4, nv_15, how='left', on=['County'])
result_5.head()
```

```python
# Merging with 2016 data, renaming columns and dropping NA values
result_6 = pd.merge(result_5, nv_16, how='left', on=['County'])
result_6 = result_6.dropna()
result_6 = result_6.rename(columns={'Violent crime_x':'2014','Violent crime_y':'2015','Violent crime':'2016'})
result_6.head()
```

```python
# Sum of all Results
sum_2010 = result_6['2010'].sum()
sum_2011 = result_6['2011'].sum()
sum_2012 = result_6['2012'].sum()
sum_2013 = result_6['2013'].sum()
sum_2014 = result_6['2014'].sum()
sum_2015 = result_6['2015'].sum()
sum_2016 = result_6['2016'].sum()
sum_all = [sum_2010, sum_2011, sum_2012, sum_2013, sum_2014, sum_2015, sum_2016]
```

```python
# Plotting the data
x_vals = ['2010','2011','2012','2013','2014','2015','2016']

plt.plot(x_vals, sum_all, color = 'forestgreen', linewidth = 4)
plt.title("Crime per Year in Nevada (2010-2016)", loc = 'center',fontsize = 17,fontweight = 3, color = 'black' )
plt.xlabel("Year" , fontsize = 15, fontweight = 3, color = 'black' )
plt.ylabel("Sum of Crime", fontsize = 15, fontweight= 3, color = 'black' )
sns.set(rc={'axes.facecolor':'whitesmoke', 'figure.facecolor':'lightgrey'})
plt.grid(color='gainsboro')
plt.xlim(2010,2016)
plt.savefig("Nevada_Crime_2010_2016")
```
