```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2010 data
file_path = "County_data/Oregon_county_2010.csv"
Or_10 = pd.read_csv(file_path)
Or_10 = Or_10[['County','Violent crime']]
Or_10.head()
```

```python
# 2011 data
file_path = "County_data/Oregon_county_2011.csv"
Or_11 = pd.read_csv(file_path)
Or_11 = Or_11[['County','Violent crime']]
Or_11.head()
```

```python
# 2012 data
file_path = "County_data/Oregon_county_2012.csv"
Or_12 = pd.read_csv(file_path)
Or_12 = Or_12[['County','Violent crime']]
Or_12.head()
```

```python
# 2013 data
file_path = "County_data/Oregon_county_2013.csv"
Or_13 = pd.read_csv(file_path)
Or_13 = Or_13[['County','Violent crime']]
Or_13.head()
```

```python
# 2014 data
file_path = "County_data/Oregon_county_2014.csv"
Or_14 = pd.read_csv(file_path)
Or_14 = Or_14[['County','Violent crime']]
Or_14.head()
```

```python
# 2015 data
file_path = "County_data/Oregon_county_2015.csv"
Or_15 = pd.read_csv(file_path)
Or_15 = Or_15[['County','Violent crime']]
Or_15.head()
```

```python
# 2016 data
file_path = "County_data/Oregon_county_2016.csv"
Or_16 = pd.read_csv(file_path)
Or_16 = Or_16[['County','Violent crime']]
Or_16.head()
```

```python
# Merging 2010 and 2011 data
result_1 = pd.merge(Or_10, Or_11, how='left', on=['County'])
result_1 = result_1.rename(columns={'Violent crime_x':'2010','Violent crime_y':'2011'})
result_1.head()
```

```python
# Merging with 2012 data
result_2 = pd.merge(result_1, Or_12, how='left', on=['County'])
result_2.head()
```

```python
# Merging with 2013 data
result_3 = pd.merge(result_2, Or_13, how='left', on=['County'])
result_3 = result_3.rename(columns={'Violent crime_x':'2012','Violent crime_y':'2013'})
result_3
```

```python
# Merging with 2014 data
result_4 = pd.merge(result_3, Or_14, how='left', on=['County'])
result_4 = result_4.dropna()
result_4.head()
```

```python
# Merging with 2015 data
result_5 = pd.merge(result_4, Or_15, how='left', on=['County'])
result_5.head()
```

```python
# Merging with 2016 data
result_6 = pd.merge(result_5, Or_16, how='left', on=['County'])
result_6 = result_6.dropna()
result_6 = result_6.rename(columns={'Violent crime_x':'2014','Violent crime_y':'2015','Violent crime':'2016'})
result_6.head()
```

```python
# Sum of all results
sum_2010 = result_6['2010'].sum()
sum_2011 = result_6['2011'].sum()
sum_2012 = result_6['2012'].sum()
sum_2013 = result_6['2013'].sum()
sum_2014 = result_6['2014'].sum()
sum_2015 = result_6['2015'].sum()
sum_2016 = result_6['2016'].sum()
sum_all = [sum_2010, sum_2011, sum_2012, sum_2013, sum_2014, sum_2015, sum_2016]
```

```python
# Plotting the data
x_vals = ['2010','2011','2012','2013','2014','2015','2016']

plt.plot(x_vals, sum_all, color = 'forestgreen', linewidth = 4)
plt.title("Crime per Year in Oregon (2010-2016)", loc = 'center',fontsize = 17,fontweight = 3, color = 'black' )
plt.xlabel("Year" , fontsize = 15, fontweight = 3, color = 'black' )
plt.ylabel("Sum of Crime", fontsize = 15, fontweight= 3, color = 'black' )
sns.set(rc={'axes.facecolor':'whitesmoke', 'figure.facecolor':'lightgrey'})
plt.grid(color='gainsboro')
plt.xlim(2010,2016)
plt.savefig("Oregon_Crime_2010_2016")
```
