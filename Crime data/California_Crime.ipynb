```python
# Dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
# Crime 2010 
ca_2010 = "County_data/California_county_2010.csv"
ca_10 = pd.read_csv(ca_2010)
ca_10 = ca_10[['County','Violent crime']]
ca_10.head()
```

```python
# Crime 2011
ca_2011 = "County_data/California_county_2011.csv"
ca_11 = pd.read_csv(ca_2011)
ca_11 = ca_11[['County','Violent crime']]
ca_11.head()
```

```python
# Crime 2012
ca_2012 = "County_data/California_county_2012.csv"
ca_12 = pd.read_csv(ca_2012)
ca_12 = ca_12[['County','Violent crime']]
ca_12.head()
```

```python
# Crime 2013
ca_2013 = "County_data/California_county_2013.csv"
ca_13 = pd.read_csv(ca_2013)
ca_13 = ca_13[['County','Violent crime']]
ca_13.head()
```

```python
# Crime 2014
ca_2014 = "County_data/California_county_2014.csv"
ca_14 = pd.read_csv(ca_2014)
ca_14 = ca_14[['County','Violent crime']]
ca_14.head()
```

```python
# Crime 2015
ca_2015 = "County_data/California_county_2015.csv"
ca_15 = pd.read_csv(ca_2015)
ca_15 = ca_15[['County','Violent crime']]
ca_15.head()
```

```python
# Crime 2016
ca_2016 = "County_data/California_county_2016.csv"
ca_16 = pd.read_csv(ca_2016)
ca_16 = ca_16[['County','Violent crime']]
ca_16.head()
```

```python
# Merging 2010-2011 and renaming columns
ca10_11 = pd.merge(ca_10, ca_11, how='left', on=['County'])
ca10_11 = ca10_11.rename(columns={'Violent crime_x':'2010','Violent crime_y':'2011'})
ca10_11.head()
```

```python
# Merging 2011-2012
merge_12 = pd.merge(ca10_11, ca_12, how='left', on=['County'])
merge_12.head()
```

```python
# Merging 2012-2013 and renaming the columns
merge_13 = pd.merge(merge_12, ca_13, how='left', on=['County'])
merge_13 = merge_13.rename(columns={'Violent crime_x':'2012','Violent crime_y':'2013'})
merge_13.head()
```

```python
# Merging 2013-2014
merge_14 = pd.merge(merge_13, ca_14, how='left', on=['County'])
merge_14.head()
```

```python
# Merging 2014-2015 and renaming the columns
merge_15 = pd.merge(merge_14, ca_15, how='left', on=['County'])
merge_15 = merge_15.rename(columns={'Violent crime_x':'2014','Violent crime_y':'2015'})
merge_15.head()
```

```python
# Merging 2015-2016 and renaming the columns
merge_16 = pd.merge(merge_15, ca_16, how='left', on=['County'])
merge_16 = merge_16.rename(columns={'Violent crime':'2016'})
merge_16.head()
```

```python
# Drop NA Values
result = merge_16.dropna()
result.head()
```

```python
# Summing all the values for charting
sum_2010 = result['2010'].sum()
sum_2011 = result['2011'].sum()
sum_2012 = result['2012'].sum()
sum_2013 = result['2013'].sum()
sum_2014 = result['2014'].sum()
sum_2015 = result['2015'].sum()
sum_2016 = result['2016'].sum()
sum_all = [sum_2010, sum_2011, sum_2012, sum_2013, sum_2014, sum_2015, sum_2016]
```

```python
# Plotting the data
x_vals = ['2010','2011','2012','2013','2014','2015','2016']

plt.plot(x_vals, sum_all)
plt.title("Crime per Year in California (2010-2016)")
plt.xlabel("Year")
plt.ylabel("Sum of Crime")
plt.xlim(2010,2016)
plt.savefig("California_Crime_2010_2016")
sns.set(rc={'axes.facecolor':'paleturquoise', 'figure.facecolor':'paleturquoise'})
```
