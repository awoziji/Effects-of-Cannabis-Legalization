```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2010 Data
mi_10_path = "County_data/Michigan_county_2010.csv"
mi_10 = pd.read_csv(mi_10_path)
mi_10 = mi_10[['County','Violent crime']]
mi_10.head()
```

```python
# 2011 Data
mi_11_path = "County_data/Michigan_county_2011.csv"
mi_11 = pd.read_csv(mi_11_path)
mi_11 = mi_11[['County','Violent crime']]
mi_11.head()
```

```python
# 2012 Data
file_path = "County_data/Michigan_county_2012.csv"
mi_12 = pd.read_csv(file_path)
mi_12 = mi_12[['County','Violent crime']]
mi_12.head()
```

```python
# 2013 Data
file_path = "County_data/Michigan_county_2013.csv"
mi_13 = pd.read_csv(file_path)
mi_13 = mi_13[['County','Violent crime']]
mi_13.head()
```

```python
# 2014 data
file_path = "County_data/Michigan_county_2014.csv"
mi_14 = pd.read_csv(file_path)
mi_14 = mi_14[['County','Violent crime']]
mi_14.head()
```

```python
# 2015 Data
file_path = "County_data/Michigan_county_2015.csv"
mi_15 = pd.read_csv(file_path)
mi_15 = mi_15[['County','Violent crime']]
mi_15.head()
```

```python
# 2016 data
file_path = "County_data/Michigan_county_2016.csv"
mi_16 = pd.read_csv(file_path)
mi_16 = mi_16[['County','Violent crime']]
mi_16.head()
```

```python
# Merge 2010-2011 data
result_1 = pd.merge(mi_10, mi_11, how='left', on=['County'])
result_1 = result_1.rename(columns={'Violent crime_x':'2010','Violent crime_y':'2011'})
result_1.head()
```

```python
# Merging with 2012 data
result_2 = pd.merge(result_1, mi_12, how='left', on=['County'])
result_2.head()
```

```python
# Merging with 2013 data
result_3 = pd.merge(result_2, mi_13, how='left', on=['County'])
result_3 = result_3.rename(columns={'Violent crime_x':'2012','Violent crime_y':'2013'})
result_3.head()
```

```python
# Merging with 2014 data
result_4 = pd.merge(result_3, mi_14, how='left', on=['County'])
result_4 = result_4.dropna()
result_4.head()
```

```python
# Merging with 2015 data
result_5 = pd.merge(result_4, mi_15, how='left', on=['County'])
result_5.head()
```

```python
# Merging with 2016 data
result_6 = pd.merge(result_5, mi_16, how='left', on=['County'])
result_6 = result_6.dropna()
result_6 = result_6.rename(columns={'Violent crime_x':'2014','Violent crime_y':'2015','Violent crime':'2016'})
result_6.head()
```

```python
# Sum of all values
sum_2010 = result_6['2010'].sum()
sum_2011 = result_6['2011'].sum()
sum_2012 = result_6['2012'].sum()
sum_2013 = result_6['2013'].sum()
sum_2014 = result_6['2014'].sum()
sum_2015 = result_6['2015'].sum()
sum_2016 = result_6['2016'].sum()
sum_all = [sum_2010, sum_2011, sum_2012, sum_2013, sum_2014, sum_2015, sum_2016]
```

```python
# Plotting the data
x_vals = ['2010','2011','2012','2013','2014','2015','2016']

plt.plot(x_vals, sum_all)
plt.title("Crime per Year in Michigan (2010-2016)")
plt.xlabel("Year")
plt.ylabel("Sum of Crime")
plt.xlim(2010,2016)
plt.savefig("Michigan_Crime_2010_2016")
sns.set(rc={'axes.facecolor':'paleturquoise', 'figure.facecolor':'paleturquoise'})
```
