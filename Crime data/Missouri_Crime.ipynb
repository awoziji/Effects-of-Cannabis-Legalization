```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2010 data
mo_10_path = "County_data/Missouri_county_2010.csv"
mo_10 = pd.read_csv(mo_10_path)
mo_11 = mo_10[['County','Violent crime']]
mo_10.head()
```

```python
# 2011 data
mo_11_path = "County_data/Missouri_county_2011.csv"
mo_11 = pd.read_csv(mo_11_path)
mo_11 = mo_11[['County','Violent crime']]
mo_11.head()
```

```python
# 2012 data
file_path = "County_data/Missouri_county_2012.csv"
mo_12 = pd.read_csv(file_path)
mo_12 = mo_12[['County','Violent crime']]
mo_12.head()
```

```python
# 2013 data
file_path = "County_data/Missouri_county_2013.csv"
mo_13 = pd.read_csv(file_path)
mo_13 = mo_13[['County','Violent crime']]
mo_13.head()
```

```python
# 2014 data
file_path = "County_data/Missouri_county_2014.csv"
mo_14 = pd.read_csv(file_path)
mo_14 = mo_14[['County','Violent crime']]
mo_14.head()
```

```python
# 2015 data
file_path = "County_data/Missouri_county_2015.csv"
mo_15 = pd.read_csv(file_path)
mo_15 = mo_15[['County','Violent crime']]
mo_15.head()
```

```python
# 2016 data
file_path = "County_data/Missouri_county_2016.csv"
mo_16 = pd.read_csv(file_path)
mo_16 = mo_16[['County','Violent crime']]
mo_16.head()
```

```python
# Merge 2010-2011 data
result_1 = pd.merge(mo_10, mo_11, how='left', on=['County'])
result_1 = result_1.rename(columns={'Violent crime_x':'2010','Violent crime_y':'2011'})
result_1.head()
```

```python
result_2 = pd.merge(result_1, mo_12, how='left', on=['County'])
result_2.head()
```

```python
result_3 = pd.merge(result_2, mo_13, how='left', on=['County'])
result_3 = result_3.rename(columns={'Violent crime_x':'2012','Violent crime_y':'2013'})
result_3
```

```python
result_4 = pd.merge(result_3, mo_14, how='left', on=['County'])
result_4 = result_4.dropna()
result_4 
```

```python
result_5 = pd.merge(result_4, mo_15, how='left', on=['County'])
result_5.head()
```

```python
result_6 = pd.merge(result_5, mo_16, how='left', on=['County'])
result_6 = result_6.dropna()
result_6 = result_6.rename(columns={'Violent crime_x':'2014','Violent crime_y':'2015','Violent crime':'2016'})
result_6
```

```python
# Sum of all results
sum_2010 = result_6['2010'].sum()
sum_2011 = result_6['2011'].sum()
sum_2012 = result_6['2012'].sum()
sum_2013 = result_6['2013'].sum()
sum_2014 = result_6['2014'].sum()
sum_2015 = result_6['2015'].sum()
sum_2016 = result_6['2016'].sum()
sum_all = [sum_2010, sum_2011, sum_2012, sum_2013, sum_2014, sum_2015, sum_2016]
```

```python
# Plotting the data
x_vals = ['2010','2011','2012','2013','2014','2015','2016']

plt.plot(x_vals, sum_all)
plt.title("Crime per Year in Missouri(2010-2016)")
plt.xlabel("Year")
plt.ylabel("Sum of Crime")
plt.xlim(2010,2016)
plt.savefig("Missouri_Crime_2010_2016")
sns.set(rc={'axes.facecolor':'paleturquoise', 'figure.facecolor':'paleturquoise'})
```
