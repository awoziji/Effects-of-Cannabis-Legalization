```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col_10_path = "County_data/Colorado_county_2010.csv"
col_10 = pd.read_csv(col_10_path)
col_10 = col_10[['County','Violent crime']]
col_10.head()
```

```python
col_11_path = "County_data/Colorado_county_2011.csv"
col_11 = pd.read_csv(col_11_path)
col_11 = col_11[['County','Violent crime']]
col_11.head()
```

```python
file_path = "County_data/Colorado_county_2012.csv"
col_12 = pd.read_csv(file_path)
col_12 = col_12[['County','Violent crime']]
col_12.head()
```

```python
file_path = "County_data/Colorado_county_2013.csv"
col_13 = pd.read_csv(file_path)
col_13 = col_13[['County','Violent crime']]
col_13.head()
```

```python
file_path = "County_data/Colorado_county_2014.csv"
col_14 = pd.read_csv(file_path)
col_14 = col_14[['County','Violent crime']]
col_14.head()
```

```python
file_path = "County_data/Colorado_county_2015.csv"
col_15 = pd.read_csv(file_path)
col_15 = col_15[['County','Violent crime']]
col_15.head()
```

```python
file_path = "County_data/Colorado_county_2016.csv"
col_16 = pd.read_csv(file_path)
col_16 = col_16[['County','Violent crime']]
col_16.head()
```

```python
result_1 = pd.merge(col_10, col_11, how='left', on=['County'])
result_1 = result_1.rename(columns={'Violent crime_x':'2010','Violent crime_y':'2011'})
result_1
```

```python
result_2 = pd.merge(result_1, col_12, how='left', on=['County'])
result_2
```

```python
result_3 = pd.merge(result_2, col_13, how='left', on=['County'])
result_3 = result_3.rename(columns={'Violent crime_x':'2012','Violent crime_y':'2013'})
result_3
```

```python
result_4 = pd.merge(result_3, col_14, how='left', on=['County'])
result_4 = result_4.dropna()
result_4 
```

```python
result_5 = pd.merge(result_4, col_15, how='left', on=['County'])
result_5
```

```python
result_6 = pd.merge(result_5, col_16, how='left', on=['County'])
result_6 = result_6.dropna()
result_6 = result_6.rename(columns={'Violent crime_x':'2014','Violent crime_y':'2015','Violent crime':'2016'})
result_6
```

```python
sum_2010 = result_6['2010'].sum()
sum_2011 = result_6['2011'].sum()
sum_2012 = result_6['2012'].sum()
sum_2013 = result_6['2013'].sum()
sum_2014 = result_6['2014'].sum()
sum_2015 = result_6['2015'].sum()
sum_2016 = result_6['2016'].sum()
sum_all = [sum_2010, sum_2011, sum_2012, sum_2013, sum_2014, sum_2015, sum_2016]
```

```python
#x values for the chart
x_vals = ['2010','2011','2012','2013','2014','2015','2016']

plt.plot(x_vals, sum_all)
sns.set_style("darkgrid")
plt.savefig("Colorado_Crime_2010_2016")
```
