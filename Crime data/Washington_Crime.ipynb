```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Crime 2010 
file_path = "County_data/Washington_county_2010.csv"
Wa_10 = pd.read_csv(file_path)
Wa_10 = Wa_10[['County','Violent crime']]
Wa_10.head()
```

```python
# Crime 2011
file_path = "County_data/Washington_county_2011.csv"
Wa_11 = pd.read_csv(file_path)
Wa_11 = Wa_11[['County','Violent crime']]
Wa_11.head()
```

```python
# Crime 2012 data
file_path = "County_data/Washington_county_2012.csv"
Wa_12 = pd.read_csv(file_path)
Wa_12 = Wa_12[['County','Violent crime']]
Wa_12.head()
```

```python
# Crime 2013 data
file_path = "County_data/Washington_county_2013.csv"
Wa_13 = pd.read_csv(file_path)
Wa_13 = Wa_13[['County','Violent crime']]
Wa_13.head()
```

```python
# Crime 2014 data
file_path = "County_data/Washington_county_2014.csv"
Wa_14 = pd.read_csv(file_path)
Wa_14 = Wa_14[['County','Violent crime']]
Wa_14.head()
```

```python
# Crime 2015 data
file_path = "County_data/Washington_county_2015.csv"
Wa_15 = pd.read_csv(file_path)
Wa_15 = Wa_15[['County','Violent crime']]
Wa_15.head()
```

```python
# Crime 2016 data
result_1 = pd.merge(Wa_10, Wa_11, how='left', on=['County'])
result_1 = result_1.rename(columns={'Violent crime_x':'2010','Violent crime_y':'2011'})
result_1.head()
```

```python
# Merging 2010-2011 and renaming columns
result_2 = pd.merge(result_1, Wa_12, how='left', on=['County'])
result_2.head()
```

```python
# Merging with 2012 and renaming columns
result_3 = pd.merge(result_2, Wa_13, how='left', on=['County'])
result_3 = result_3.rename(columns={'Violent crime_x':'2012','Violent crime_y':'2013'})
result_3.head()
```

```python
# Merging with 2014 data
result_4 = pd.merge(result_3, Wa_14, how='left', on=['County'])
result_4 = result_4.dropna()
result_4.head()
```

```python
# Merging with 2015 data
result_5 = pd.merge(result_4, Wa_15, how='left', on=['County'])
result_5 = result_5.dropna()
result_5 = result_5.rename(columns={'Violent crime_x':'2014','Violent crime_y':'2015'})
result_5.head()
```

```python
# Sum of all totals 
sum_2010 = result_5['2010'].sum()
sum_2011 = result_5['2011'].sum()
sum_2012 = result_5['2012'].sum()
sum_2013 = result_5['2013'].sum()
sum_2014 = result_5['2014'].sum()
sum_2015 = result_5['2015'].sum()
sum_all = [sum_2010, sum_2011, sum_2012, sum_2013, sum_2014, sum_2015]
```

```python
# Plotting the data
x_vals = ['2010','2011','2012','2013','2014','2015']

plt.plot(x_vals, sum_all, color = 'forestgreen', linewidth = 4)
plt.title("Crime per Year in Washington (2010-2015)", loc = 'center',fontsize = 17,fontweight = 3, color = 'black' )
plt.xlabel("Year" , fontsize = 15, fontweight = 3, color = 'black' )
plt.ylabel("Sum of Crime", fontsize = 15, fontweight= 3, color = 'black' )
sns.set(rc={'axes.facecolor':'whitesmoke', 'figure.facecolor':'lightgrey'})
plt.grid(color='gainsboro')
plt.xlim(2010,2016)
plt.savefig("Washington_Crime_2010_2015")
```
